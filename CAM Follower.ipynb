```python
# WRITE A PROGRAM TO DRAW KNIFE-EDGE CAM FOLLOWER PERFORMING SHM
# USING PYTHON

# CAM Follower- A cam follower, also known as a track follower,[1] is a specialized type of roller or needle bearing designed
# to follow cam lobe profiles. Cam followers come in a vast array of different configurations, however the most defining
# characteristic is how the cam follower mounts to its mating part; stud style cam followers use a stud while the yoke style 
# has a hole through the middle.

# Libraries & Dependencies used- Matplotlib, numpy, math class

import numpy as np
import matplotlib.pyplot as plt
import math

N=300

H=int(input('Enter CAM Lift Value'))
Out=int(input('Enter Out stroke value'))
Dwell=int(input('Enter Dwell value'))
Return=int(input('Enter the Return stroke value'))
Offset= int(input ('Enter Offset value'))

first= 90-Offset
second= first+Out
third= second+Dwell
fourth= third+Return

h= H*0.5
A=[]
B=[]
A3=[]
B3=[]



Theta1= np.linspace(0, Out, N)
for x in np.linspace(270, 90, N):
    a1= math.sin(math.radians(x))
    a2= h*a1
    if a2<0:
        a= a2+h
        a3= a+H
    elif a2>=0:
        a=a2+h
        a3= a+H
    A.append(a)
    A3.append(a3)
    
Theta2=np.linspace(Out+Dwell, Out+Dwell+Return, N)
for x in np.linspace(90,270 , N):
    b1=math.sin(math.radians(x))
    b2=h*b1
    if b2<0:
        b= b2+h
        b3= b+H
    elif b2>=0:
        b= b2+h
        b3= b+H
    B.append(b)
    B3.append(b3)
    
cc1= A3*np.cos(np.radians(np.linspace(first, second, N)))
cs1= A3*np.sin(np.radians(np.linspace(first, second, N)))

cc2= 2*H*np.cos(np.radians(np.linspace(second, third, N)))
cs2= 2*H*np.sin(np.radians(np.linspace(second, third, N)))

cc3= B3*np.cos(np.radians(np.linspace( third, fourth, N)))
cs3= B3*np.sin(np.radians(np.linspace( third, fourth, N)))

cc4= H*np.cos(np.radians(np.linspace(fourth, 360, N)))
cs4= H*np.sin(np.radians(np.linspace(fourth, 360, N)))

cc5= H*np.cos(np.radians(np.linspace(0, first, N)))
cs5= H*np.sin(np.radians(np.linspace(0, first, N)))

cc6= 10*np.cos(np.radians(np.linspace(0, 360, N)))
cs6= 10*np.sin(np.radians(np.linspace(0, 360, N)))

fig= plt.subplots(figsize= (12,5))
plt.grid(True)
plt.xlabel('Theta')
plt.ylabel('Lift')
plt.xticks(np.arange(0, Out+Dwell+Return+10, step=10))
plt.yticks(np.arange(0, H+5, step=5))
plt.plot(Theta1, A,Theta2, B)

fig= plt.subplots(figsize= (12,10))
plt.grid(True)
plt.plot(cc1, cs1, cc2, cs2, cc3, cs3, cc4, cs4, cc5, cs5, cc6, cs6, color= 'indigo')
plt.fill_between(cc1, cs1, color= 'pink')
plt.fill_between(cc2, cs2, color= 'pink')
plt.fill_between(cc3, cs3, color= 'pink')
plt.fill_between(cc4, cs4, color= 'pink')
plt.fill_between(cc5, cs5, color= 'pink')
plt.plot(cc6, cs6, color= 'indigo')
plt.fill_between(cc6, cs6, color= 'pink')



```
